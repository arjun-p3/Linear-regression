```python
#Creating Dataset

import pandas as pd
import numpy as np


W=6.78
B=5

x_train = np.arange(0,100)

y_train= W * x_train + B
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
#test_train_split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.3)
```

```python
#creating model



class Linear:
  def __init__(self,LR_):
    self.Weight = 0
    self.Bias = 0
    self.LR = LR_
    self.Cost = []

  def predict(self,X):
    W=self.Weight
    B=self.Bias
    return( X * W + B)

  def cost(self,X_train,Y_train):

    prediction = np.array( list( map( self.predict,X_train )))
    error = (prediction - Y_train)**2
    cost = 1/(2*len(X_train)) * np.sum(error)
    return cost

  def gradient_descent(self,epoch,X_train,Y_train):


    for x in range(epoch):

      prediction = np.array( list( map( self.predict,X_train )))
      error = np.subtract( Y_train, prediction)

      D_W = (-2/len(X_train)) * np.sum( np.dot(X_train,error))
      D_B = (-2/len(X_train)) * np.sum( error)

      self.Weight -= self.LR * D_W
      self.Bias -= self.LR * D_B

      cost = self.cost(X_train,Y_train)
      self.Cost.append(cost)

      if(x % 20 == 0):
        print(f"epoch = {x}     ||   cost={cost}")
```

```python
#calling model

model_1 = Linear(0.001)
```

```python
#training

model_1.gradient_descent(1000,X_train,Y_train)
```

```python

#Cost function

plt.cla()

Iteration = np.arange(0,200)
Y_axis=[]
for y in range(0,200):
  Y_axis.append(model_1.Cost[y])

cost = np.array(Y_axis)

plt.plot(Iteration, cost)

plt.xlabel("Iteration")
plt.ylabel("Cost")

plt.show()
```

```python
#lowering learning rate

model_2=Linear(0.0001)
```

```python
#training

model_2.gradient_descent(700,X_train,Y_train)
```

```python
#Cost function

Iteration = np.arange(0,700)
Y_axis=[]
for y in range(0,700):
  Y_axis.append(model_2.Cost[y])

cost = np.array(Y_axis)

plt.plot(Iteration, cost)

plt.xlabel("Iteration")
plt.ylabel("Cost")

plt.show()

print(model_2.Weight, model_2.Bias)
```

```python
#evaluating model on test set

predictions=[]

for x in range(0,len(X_test)):

  predictions.append(model_2.predict(X_test[x]))

Y_pred=np.array(predictions)

plt.plot(X_test,Y_test,'r',label="Test Data")
plt.plot(X_test,Y_pred,'b',label="Predicted Values")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

```python
#calculating error

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(Y_test,Y_pred)
mse = mean_squared_error(Y_test,Y_pred)
mae = mean_absolute_error(Y_test,Y_pred)

print(f"r2 score: {r2}\n mse: {mse}\n mae: {mae}")
```
